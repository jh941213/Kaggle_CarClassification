```python
from ultralytics import YOLO
import torch

# Load a model
model = YOLO("/kaggle/input/d/yuseokjung/counting-and-categorizing-vehicles-in-surveillance/0303_v8m_e159_best.pt") 
source = '/kaggle/input/counting-and-categorizing-vehicles-in-surveillance/test/jpg'
# source = '/kaggle/input/vehicles'
```

```python
results = model.predict(source,conf=0.6)
```

```python
submission = {"filename" : [], "result":[]}

for i in range(len(results)):
    filename = results[i].path.split("/")[-1]
    result = [0]*12
    for j in range(len(results[i].boxes.boxes)):
        box_index = int(results[i].boxes.boxes[j][-1])
        result[box_index] += 1
    result = [sum(result)] + result
    result_str = ','.join(str(val) for val in result)
    
    submission["filename"].append(filename)
    submission["result"].append(result_str)
    
submission
```

```python
import csv
with open('submission.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['filename', 'result'])
    for i in range(len(submission['filename'])):
        writer.writerow([submission['filename'][i], submission['result'][i]])
```
